```python
import sys
print(sys.version_info)
from io import BytesIO

import math
from sqlalchemy import create_engine
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

print(matplotlib.__version__)
from sklearn.impute import KNNImputer

from sklearn.preprocessing import StandardScaler
```

```python
localhost = {'user': 'postgres', 'password': 'postgres', 'host': 'localhost', 'port': 5432, 'db': 'fiadb'}
params = 'postgresql://{0}:{1}@{2}:{3}/{4}'
engine = create_engine(params.format(localhost['user'], localhost['password'], localhost['host'], localhost['port'], localhost['db']))
# geom_sql = """select distinct grid_id, grid_geom from fs_fiadb.pergrid"""
pergrid_base = """select distinct * from predictor.pergrid_base"""
pergrid_base_df = pd.read_sql(pergrid_base, engine)
print(pergrid_base_df.shape)
pergrid_base_df_na_remove = pergrid_base_df.dropna()
pergrid_base_df_na_remove.columns = map(str.upper, pergrid_base_df_na_remove.columns)

print(pergrid_base_df_na_remove.columns)
```

```python
pergrid_base_df_na_remove.columns
```

```python
# pergrid_base_encoded=pd.get_dummies(pergrid_base_df_na_remove, columns=["hydrogroup"])
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
pergrid_base_df_na_remove.loc[: ,'SHG'] = number.fit_transform(pergrid_base_df_na_remove['SHG'])
```

```python
PRED_VAR= ['AET', 'AI', 'ART', 'EWD', 'FA', 'MAP', 'MAT', 'MPDQ',
       'MTCQ', 'PET', 'PSN', 'RA', 'RMAP', 'RMAT', 'TSN', 'MFDF', 'ALT',
       'SHG', 'MTWQ', 'WA']

VAR= ['AET', 'AI', 'ART', 'EWD', 'FA', 'MAP', 'MAT', 'MPDQ',
       'MTCQ', 'PET', 'PSN', 'RA', 'RMAP', 'RMAT', 'TSN', 'MFDF', 'ALT',
       'SHG', 'MTWQ', 'WA', 'TSR']
```

```python
len(PRED_VAR)
```

```python
X = pergrid_base_df_na_remove[PRED_VAR]

# standarize predictors
ss = StandardScaler(with_mean=True, with_std=True)
X_std = ss.fit_transform(X.values)

Xstd=pd.DataFrame(data=X_std[0:,0:],
                index=X.index,
                columns=PRED_VAR)
y = pergrid_base_df_na_remove['TSR']
```

```python
Xstd.head()
```

### pearson correlation

```python
pearson_cor_pd = pd.DataFrame()
from scipy.stats import pearsonr
for var in PRED_VAR:
    new_row = {'feature': var, 
               'pearson_corr': abs(pearsonr(X[var], y)[0]), 
               'p_value': pearsonr(X[var], y)[1]}
    pearson_cor_pd = pearson_cor_pd.append(new_row, ignore_index=True)
pearson_cor_pd.sort_values(by=['pearson_corr'], ascending=False)
```

## mutual_info feature selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression, mutual_info_regression
```

```python
# define feature selection as mutual info
mutual_info_best_classifier = SelectKBest(score_func=mutual_info_regression, k=20)

fit = mutual_info_best_classifier.fit(Xstd, y)
mask = mutual_info_best_classifier.get_support() #list of booleans
sel_features_mutual_info = [] # The list of your K best features

for bool, feature in zip(mask, PRED_VAR):
    if bool:
        sel_features_mutual_info.append(feature)
        
mutual_info_score_pd = pd.DataFrame(
    {'feature': sel_features_mutual_info,
     'score': fit.scores_.tolist()
    })
```

```python
mutual_info_score_pd.sort_values(by=['score'], ascending=False)
```

```python
# plot the scores
plt.bar(mutual_info_score_pd['feature'], mutual_info_score_pd['score'])
plt.show()
```

## f_regression feature selection

```python
# define feature selection as f regression
f_regression_best_classifier = SelectKBest(score_func=f_regression, k=20)

fit = f_regression_best_classifier.fit(Xstd, y)
mask = f_regression_best_classifier.get_support() #list of booleans
sel_features_f_regression = [] # The list of your K best features

for bool, feature in zip(mask, PRED_VAR):
    if bool:
        sel_features_f_regression.append(feature)

f_regression_score_pd = pd.DataFrame(
    {'feature': sel_features_f_regression,
     'score': fit.scores_.tolist()
    })
```

```python
f_regression_score_pd.sort_values(by=['score'], ascending=False)
```

```python
# plot the scores
plt.bar(f_regression_score_pd['feature'], f_regression_score_pd['score'])
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import LinearRegression
```

```python
# define the evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
```

```python
# define the pipeline to evaluate
# model = LinearRegression()
from sklearn.ensemble import RandomForestRegressor
rf_exp = RandomForestRegressor(n_estimators= 200, 
                               min_samples_split = 10,
                               min_samples_leaf = 5,
                               max_features = 'sqrt',
                               max_depth = 68,
                               bootstrap = True)
# model = RandomForestRegressor(n_estimators = 10, random_state = 42, bootstrap=True)
fs = SelectKBest(score_func=f_regression)
pipeline = Pipeline(steps=[('sel',fs), ('lr', rf_exp)])
```

```python
# define the grid
grid = dict()
grid['sel__k'] = [i for i in range(1, len(PRED_VAR)+1)]
```

```python
# define the grid search
search = GridSearchCV(pipeline, grid, scoring='neg_mean_absolute_error', n_jobs=-1, cv=cv)
# perform the search
results = search.fit(Xstd, y)
```

```python
# summarize best
print('Best MAE: %.3f' % results.best_score_)
print('Best Config: %s' % results.best_params_)
```

```python
# summarize all
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```
