```python
localhost = {'user': 'postgres', 'password': 'postgres', 'host': 'localhost', 'port': 5432, 'db': 'fiadb'}
params = 'postgresql://{0}:{1}@{2}:{3}/{4}'
engine = create_engine(params.format(localhost['user'], localhost['password'], localhost['host'], localhost['port'], localhost['db']))
pergrid_base = """select distinct * from predictor.pergrid_base"""
pergrid_base_df = pd.read_sql(pergrid_base, engine)
```

```python
var =  ['aet', 'ai', 'art', 'ewd', 'fa', 'map', 'mat', 'mpdq',
       'mtcq', 'pet', 'psn', 'ra', 'rmap', 'rmat', 'tsn', 'mfdf', 'alt','shg', 'mtwq', 'wa', 'tsr']
VAR= ['AET', 'AI', 'ART', 'EWD', 'FA', 'MAP', 'MAT', 'MPDQ',
       'MTCQ', 'PET', 'PSN', 'RA', 'RMAP', 'RMAT', 'TSN', 'MFDF', 'ALT','SHG', 'MTWQ', 'WA', 'TSR']

pergrid_base_df = pergrid_base_df[var]
```

```python
pergrid_base_df2 = pergrid_base_df.copy()
pergrid_base_df2.shg = pergrid_base_df2.shg.replace({None: ""})
pergrid_base_df2.head()
```

### calcualte spearman correlation coefficient

```python
from scipy.stats import spearmanr 

corr = spearmanr(pergrid_base_df2).correlation
corr_sort_abs = pd.DataFrame(corr, columns=var)['tsr'].abs()
corr_sort_abs_df = pd.DataFrame(
    {'feature': VAR,
    'correlation': corr_sort_abs.tolist()})
corr_sort_abs_df.drop(corr_sort_abs_df.tail(1).index,inplace=True)
corr_sort_abs_df.head()
```

### calculate permutation importance

```python
categorical_columns = ['shg']
numerical_columns = ['aet', 'ai', 'art', 'ewd', 'fa', 'map', 'mat', 'mpdq',
       'mtcq', 'pet', 'psn', 'ra', 'rmap', 'rmat', 'tsn', 'mfdf', 'alt',
       'mtwq', 'wa']
```

```python
X = pergrid_base_df2[categorical_columns + numerical_columns]
y = pergrid_base_df2['tsr']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

categorical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
numerical_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='mean'))
])
```

```python
preprocessing = ColumnTransformer(
    [('cat', categorical_pipe, categorical_columns),
     ('num', numerical_pipe, numerical_columns)])
```

```python
rf_classifer = RandomForestRegressor(n_estimators = 10, random_state = 42, 
                                         min_samples_leaf=10, 
                                         bootstrap=True)
rf = Pipeline([
    ('preprocess', preprocessing),
    ('classifier', rf_classifer)
])
rf.fit(X_train, y_train)
```

```python
print("RF train accuracy: %0.3f" % rf.score(X_train, y_train))
print("RF test accuracy: %0.3f" % rf.score(X_test, y_test))
```

```python
from sklearn.inspection import permutation_importance

result = permutation_importance(rf, X_test, y_test, n_repeats=10, random_state=0)
perm_sorted_idx = result.importances_mean.argsort()
```

```python
type(X_train.columns[perm_sorted_idx])
```

```python
perm_importance = pd.DataFrame(
    {'feature': X_train.columns[perm_sorted_idx],
    'perm_importance': result.importances_mean[perm_sorted_idx].T})
```

```python
perm_importance['feature'] = perm_importance['feature'].str.upper() 
perm_importance.head()
```

```python
Xstd.head()
```

```python
# X_train, X_test, y_train, y_test = train_test_split(Xstd, y_true, test_size=0.2, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(Xstd, y_true, test_size=0.1, random_state=0)
```

```python
rf_exp = RandomForestRegressor(n_estimators= 200, 
                               min_samples_split = 10,
                               min_samples_leaf = 5,
                               max_features = 'sqrt',
                               max_depth = 68,
                               bootstrap = True)
rf_exp.fit(X_train, y_train)
```

```python
feature_list = list(X_train.columns)
# Get numerical feature importances
importances = list(rf_exp.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 6)) for feature, importance in zip(feature_list, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)

# List of features sorted from most to least important
sorted_importances = [importance[1] for importance in feature_importances]

sorted_features = [importance[0].upper() for importance in feature_importances]

# Cumulative importances
cumulative_importances = np.cumsum(sorted_importances)
```

```python
rf_importance = pd.DataFrame(
    {'feature': sorted_features,
    'rf_importance': sorted_importances,
    'cum_rf_importance': cumulative_importances
    })
```

```python
rf_importance.head()
```
