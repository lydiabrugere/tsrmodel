```python
from sqlalchemy import create_engine
import sys
import pandas as pd
import matplotlib.pyplot as plt
```

```python
localhost = {'user': 'postgres', 'password': 'postgres', 'host': 'localhost', 'port': 5432, 'db': 'fiadb'}
params = 'postgresql://{0}:{1}@{2}:{3}/{4}'
engine = create_engine(params.format(localhost['user'], localhost['password'], localhost['host'], localhost['port'], localhost['db']))
# geom_sql = """select distinct grid_id, grid_geom from fs_fiadb.pergrid"""
lasso_y_test = """select * from predictor.lasso_y_test"""
lasso_y_test_pd = pd.read_sql(lasso_y_test, engine)

glm_y_test = """select * from predictor.glm_y_test"""
glm_y_test_pd = pd.read_sql(glm_y_test, engine)

rf_y_test = """select * from predictor.rf_y_test"""
rf_y_test_pd = pd.read_sql(rf_y_test, engine)

grnn_y_test = """select * from predictor.grnn_y_test"""
grnn_y_test_pd = pd.read_sql(grnn_y_test, engine)

ffnn_y_test = """select * from predictor.ffnn_y_test"""
ffnn_y_test_pd = pd.read_sql(ffnn_y_test, engine)
```

```python
lasso_y_test_pd = lasso_y_test_pd.sample(n=int(lasso_y_test_pd.shape[0]/2), random_state=1234)
```

```python
glm_y_test_pd = glm_y_test_pd.sample(n=int(glm_y_test_pd.shape[0]/2), random_state=1234)
```

```python
rf_y_test_pd = rf_y_test_pd.sample(n=int(rf_y_test_pd.shape[0]/2), random_state=1234)
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
import matplotlib.lines as mlines

rcParams['figure.figsize'] = 10, 10

fig, axes = plt.subplots(nrows = 2, ncols = 2)

axes[0][0].scatter(lasso_y_test_pd['tsr'], lasso_y_test_pd['tsr_predicted'])
axes[0][0].set_title('(a) LASSO ($R^2$) = 0.797')
axes[0][0].set_xlabel('Observed TSR')
axes[0][0].set_ylabel('Predicted TSR')
axes[0][0].grid(False)
line = mlines.Line2D([0, 1], [0, 1], color='black')
transform = axes[0][0].transAxes
line.set_transform(transform)
axes[0][0].add_line(line)

axes[0][1].scatter(rf_y_test_pd['tsr'], rf_y_test_pd['tsr_predicted'])
axes[0][1].set_title('(b) Random Forest ($R^2$) = 0.903')
axes[0][1].set_xlabel('Observed TSR')
axes[0][1].set_ylabel('Predicted TSR')
axes[0][1].grid(False)
line = mlines.Line2D([0, 1], [0, 1], color='black')
transform = axes[0][1].transAxes
line.set_transform(transform)
axes[0][1].add_line(line)

axes[1][0].scatter(grnn_y_test_pd['tsr'], grnn_y_test_pd['tsr_predicted'])
axes[1][0].set_title('(c) GRNN')
axes[1][0].set_xlabel('Observed TSR')
axes[1][0].set_ylabel('Predicted TSR')
axes[1][0].grid(False)
line = mlines.Line2D([0, 1], [0, 1], color='black')
transform = axes[1][0].transAxes
line.set_transform(transform)
axes[1][0].add_line(line)

axes[1][1].scatter(ffnn_y_test_pd['tsr'], ffnn_y_test_pd['tsr_predicted'])
axes[1][1].set_title('(d) FFNN')
axes[1][1].set_xlabel('Observed TSR')
axes[1][1].set_ylabel('Predicted TSR')
axes[1][1].grid(False)
line = mlines.Line2D([0, 1], [0, 1], color='black')
transform = axes[1][1].transAxes
line.set_transform(transform)
axes[1][1].add_line(line)

plt.tight_layout()
```
